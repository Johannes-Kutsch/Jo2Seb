````python
def plot_model_comparison(bg_color='#262626', text_color='#ffffff', baseline_color='#ff4d4d', gb_color='#4d9fff',
                          labelpad=15, title_font_size=30, font_size=20, figsize=(20, 10), ax_lw=2,
                          bar_width=0.5):
    models = ['T+1 Baseline', 'Gradient Boosting']
    r2_scores = [0.21, 0.40]
    rmse_scores = [9.97, 8.67]
    colors = [baseline_color, gb_color]

    x = np.arange(len(models))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    fig.patch.set_facecolor(bg_color)

    for ax, scores, metric, better in zip([ax1, ax2], [r2_scores, rmse_scores], ['R²', 'RMSE'], ['higher is better', 'lower is better']):
        ax.set_facecolor(bg_color)
        bars = ax.bar(x, scores, width=bar_width, color=colors)

        for bar, score in zip(bars, scores):
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_height() + max(scores) * 0.02,
                f'{score:.2f}',
                ha='center', va='bottom',
                fontsize=font_size, fontweight='bold', color=text_color
            )

        ax.set_xticks(x)
        ax.set_xticklabels(models, fontsize=font_size, color=text_color)
        ax.set_ylabel(metric, fontsize=font_size, fontweight='bold', color=text_color, labelpad=labelpad)
        #ax.yaxis.set_visible(False)
        ax.set_title(f'{metric}\n', fontsize=title_font_size * 0.7, fontweight='bold', color=text_color)
        ax.text(0.5, 1.02, f'({better})', transform=ax.transAxes,  ha='center', fontsize=font_size * 0.75, color=text_color)
        ax.tick_params(axis='both', labelsize=font_size, labelcolor=text_color, color=text_color, width=ax_lw)
        ax.grid(False)

        for side in ['bottom', 'left']:
            ax.spines[side].set_linewidth(ax_lw)
            ax.spines[side].set_color(text_color)
        for side in ['top', 'right']:
            ax.spines[side].set_visible(False)

        ax.set_ylim(0, max(scores) * 1.2)

    plt.tight_layout(w_pad=10)
    plt.show()
````

````python
plot_model_comparison(bar_width=0.6)
````
