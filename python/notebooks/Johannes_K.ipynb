# rough EDA

#### Preprocess and fill missing Values

````python
df = pd.read_csv(FilePaths.UWB_DATA / "Station_857_2018-01-01--2025-12-31.csv")
components_meta = pd.read_csv(FilePaths.UWB_METADATA / "UWB_components_metadata.csv")
````

````python
def get_component_metadata(component_id: int):
    return components_meta.loc[components_meta["component_id"] == component_id].squeeze()

def interpolate_linear(X):
    X_interp = pd.DataFrame(X).interpolate(method='linear', limit_direction='both')
    return X_interp.values

interpolator = FunctionTransformer(interpolate_linear)

component_columns = ["NO2", "SO2", "PM10"]

component_nan_interpolator = ColumnTransformer(
    transformers=[
        ("interpolate_nan", interpolator, component_columns)
    ],
    remainder="drop"
)
````

````python
preprocess_pipe = Pipeline([
    ("rename_features", ru.FeatureRenameTransformer({
        "5": get_component_metadata(5)["component_code"],
        "4": get_component_metadata(4)["component_code"],
        "1": get_component_metadata(1)["component_code"],
        "date_start": "date",
    })),
    ("date_type", ru.ValueTypeCastDatetimeTransformer("date")),
])

df_trans = preprocess_pipe.fit_transform(df)

component_pip = Pipeline([
    ("interpolate_columns", component_nan_interpolator)
])

df_trans[component_columns] = pd.DataFrame(component_pip.fit_transform(df_trans), columns=component_columns)
df_trans_daily = df_trans.set_index("date").resample("D").agg({
    "PM10": "mean"
})
df_trans_daily["date"] = pd.to_datetime(df_trans_daily.index.date)
````

#### Check if Dates are continuous

````python
df_trans = df_trans.sort_values('date').reset_index(drop=True)
date_diff = df_trans['date'].diff()
print(f"Distance between dates are: {date_diff.unique()[1:]}")
````

# Models

### Baseline Models

#### Predict Always Mean

````python
mean_pipeline = Pipeline([
    ("predict_mean", DummyRegressor(strategy="mean"))
])

pred_dummy_mean = evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], mean_pipeline, DAILY_EXPERIMENT_NAME, model_name="Baseline_Mean", walk_forward=True)
#pred_hourly = evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, model_name="Baseline_Mean")
````

````python
plot_prediction(pred_dummy_mean, "Dummy Mean")
````

#### Persistence / Naive Forcast / PM10(t+1) = PM10(t)

````python
naive_forcast_pipeline = Pipeline([
    ("predict_T+1", NaiveForcastRegressor())
])

pred_naive = evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], naive_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_Forcast (T+1)", walk_forward=True)
#pred_hourly = evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, model_name="Baseline_T+1")
````

````python
plot_prediction(pred_naive, "Naive T+1")
````

### Naive MSTL

````python
params = {
    "seasonal_periods": [7, 365],
    "trend_model": LinearRegression(),
    "resid_model": DummyRegressor(strategy="mean"),
}

naive_mstl_forcast_pipeline = Pipeline([
    ("Naive_MSTL_Regressor", MSTLRegressor(seasonal_periods=params["seasonal_periods"],trend_model=params["trend_model"], resid_model= params["resid_model"]))
])

pred_mstl_naive = evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], naive_mstl_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_MSTL_Forcast (T+1)", walk_forward=False, params=params)
````

````python
plot_prediction(pred_mstl_naive, "Naive MSTL")
````
