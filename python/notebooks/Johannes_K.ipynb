````python
from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline

import pandas as pd
import numpy as np

import ruhken_utils as ru
from ruhken_utils import explore_ts, explore_features, explore_dataframe, FeatureDropTransformer

from SKLearn.Decomposer.MSTLDecomposer import MSTLDecomposer
from SKLearn.Transformers.DropHeadNaN import DropHeadNaN
from SKLearn.Regressor.NaiveForcastRegressor import NaiveForcastRegressor
from SKLearn.Regressor.Utils import create_transformed_target_regressor
from SKLearn.Transformers.ColumnFunctionTransformer import ColumnFunctionTransformer
from SKLearn.Transformers.TemporalFeatureBuilder import TemporalFeatureBuilder
from Utils import FilePaths
from MLFlow.Evaluation import evaluate_timeseries_pipeline, plot_prediction
from MLFlow.Utils import DAILY_EXPERIMENT_NAME

from statsmodels.tsa.seasonal import MSTL
````

# Preprocess and fill missing Values

````python
df = pd.read_csv(FilePaths.UWB_DATA / "Station_857_2018-01-01--2025-12-31.csv")
components_meta = pd.read_csv(FilePaths.UWB_METADATA / "UWB_components_metadata.csv")
````

````python
def get_component_metadata(component_id: int):
    return components_meta.loc[components_meta["component_id"] == component_id].squeeze()

def interpolate_linear(X):
    X_interp = pd.DataFrame(X).interpolate(method='linear', limit_direction='both')
    return X_interp.values

interpolator = FunctionTransformer(interpolate_linear)

component_columns = ["NO2", "SO2", "PM10"]

component_nan_interpolator = ColumnTransformer(
    transformers=[("interpolate_nan", interpolator, component_columns)],
    remainder="drop"
)
````

````python
preprocess_pipe = Pipeline([
    ("rename_features", ru.FeatureRenameTransformer({
        "5": get_component_metadata(5)["component_code"],
        "4": get_component_metadata(4)["component_code"],
        "1": get_component_metadata(1)["component_code"],
        "date_start": "date",
    })),
    ("date_type", ru.ValueTypeCastDatetimeTransformer("date")),
])

df_trans = preprocess_pipe.fit_transform(df)

component_pipe = Pipeline([
    ("interpolate_columns", component_nan_interpolator)
])

df_trans[component_columns] = pd.DataFrame(component_pipe.fit_transform(df_trans), columns=component_columns)
df_trans_daily = df_trans.set_index("date").resample("D").agg({
    "PM10": "mean"
})
df_trans_daily["date"] = pd.to_datetime(df_trans_daily.index.date)
````

#### Temporal Features

````python
params = {
    "seasonal_periods": [7, 365],
    "trend_model": LinearRegression(),
}

temporal_pipe = Pipeline([
    ("log1p_transform_PM10", ColumnFunctionTransformer(["PM10"])),
    ("mstl_decomposition", MSTLDecomposer(column_name = "PM10", seasonal_periods=params["seasonal_periods"], trend_model=params["trend_model"], overwrite_column_with_residuals=False)),
    ("temporal_feature_builder", TemporalFeatureBuilder("PM10", lags = [1, 7, 8, 14, 15, 21, 22, 28, 29], rolling_windows=[2,3,4])),
    ("residual_temporal_feature_builder", TemporalFeatureBuilder("PM10_residuals", lags = [1, 7, 8, 14, 15, 21, 22, 28, 29], rolling_windows=[2,3,4])),
    ("drop_residual", FeatureDropTransformer(["PM10"])),
    ("drop_head_nan", DropHeadNaN())
])

df_temporal = temporal_pipe.fit_transform(df_trans_daily)
````

````python
explore_dataframe(df_temporal, target_feature_name="PM10_residuals", plot_data_spread=False, plot_pairs=False, plot_corr_matrix=False)
````

#### MSTL Features

````python
decompose_pm10_pipe = Pipeline([
    ("decompost_pm10", MSTLFeatureBuilder(
    column="PM10",
    seasonal_periods=[7, 365],
    drop_original=False))
])


df_decomposed = decompose_pm10_pipe.fit_transform(df_trans_daily)
````

# Models

### Baseline Models

#### Predict Always Mean

````python
mean_pipeline = Pipeline([
    ("predict_mean", DummyRegressor(strategy="mean")),
])

pred_dummy_mean = evaluate_timeseries_pipeline(X=df_trans_daily[['PM10']], y=df_trans_daily['PM10'], pipeline=mean_pipeline,
                                               experiment=DAILY_EXPERIMENT_NAME, model_name="Baseline_Mean", walk_forward=True)
````

````python
plot_prediction(pred_dummy_mean, "Dummy Mean")
````

#### Persistence / Naive Forcast / PM10(t+1) = PM10(t)

````python
naive_forcast_pipeline = Pipeline([
    ("predict_T+1", NaiveForcastRegressor())
])

pred_naive = evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], naive_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_Forcast (T+1)", walk_forward=True)
#pred_hourly = evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, model_name="Baseline_T+1")
````

````python
plot_prediction(pred_naive, "Naive T+1")
````

### Naive MSTL

````python
params = {
    "seasonal_periods": [7, 365],
    "trend_model": LinearRegression(),
    "resid_model": DummyRegressor(strategy="mean"),
}

naive_mstl_forcast_pipeline = Pipeline([
    ("log1p_transform_PM10", ColumnFunctionTransformer(["PM10"])),
    ("Naive_MSTL_Regressor", create_transformed_target_regressor(MSTLDecomposer(seasonal_periods=params["seasonal_periods"], trend_model=params["trend_model"], resid_model= params["resid_model"])))
])

pred_mstl_naive = evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], naive_mstl_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_MSTL_Forcast (T+1)", walk_forward=True, params=params)
````

````python
plot_prediction(pred_mstl_naive, "Naive MSTL")
````
