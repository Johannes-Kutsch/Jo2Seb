# rough EDA

#### Preprocess and fill missing Values

```python
df = pandas.read_csv(FilePaths.UWB_DATA / "Station_857_2018-01-01--2025-12-31.csv")
components_meta = pandas.read_csv(FilePaths.UWB_METADATA / "UWB_components_metadata.csv")

```

```python
def get_component_metadata(component_id: int):
    return components_meta.loc[components_meta["component_id"] == component_id].squeeze()

def interpolate_linear(X):
    X_interp = pd.DataFrame(X).interpolate(method='linear', limit_direction='both')
    return X_interp.values

interpolator = FunctionTransformer(interpolate_linear)

component_columns = ["NO2", "SO2", "PM10"]

component_nan_interpolator = ColumnTransformer(
    transformers=[
        ("interpolate_nan", interpolator, component_columns)
    ],
    remainder="drop"
)
```

```python
preprocess_pipe = Pipeline([
    ("rename_features", ru.FeatureRenameTransformer({
        "5": get_component_metadata(5)["component_code"],
        "4": get_component_metadata(4)["component_code"],
        "1": get_component_metadata(1)["component_code"],
        "date_start": "date",
    })),
    ("date_type", ru.ValueTypeCastDatetimeTransformer("date")),
])

df_trans = preprocess_pipe.fit_transform(df)

component_pip = Pipeline([
    ("interpolate_columns", component_nan_interpolator)
])

df_trans[component_columns] = pd.DataFrame(component_pip.fit_transform(df_trans), columns=component_columns)
```

#### Check if Dates are continuous

```python
df_trans = df_trans.sort_values('date').reset_index(drop=True)
date_diff = df_trans['date'].diff()
print(f"Distance between dates are: {date_diff.unique()[1:]}")
```

# Baseline Models

#### Functions

```python
def evaluate_timeseries_pipeline(X, y, pipeline, n_splits=5, print_evaluation=True):
    """
    Evaluates a scikit-learn pipeline on a time series using TimeSeriesSplit.

    Parameters:
    - X: pd.DataFrame, Feature matrix
    - y: pd.Series, Target variable
    - pipeline: sklearn Pipeline or estimator with fit/predict
    - n_splits: int, number of splits for TimeSeriesSplit
    - print_evaluation: bool, whether to print MAE/MSE per split

    Returns:
    - results: list of dicts with split info and metrics
    """
    tscv = TimeSeriesSplit(n_splits=n_splits)
    results = []

    for i, (train_index, test_index) in enumerate(tscv.split(X)):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        pipeline.fit(X_train, y_train)

        y_pred = pipeline.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))

        if print_evaluation:
            print(f"Split {i+1}: Train {len(train_index)}, Test {len(test_index)}")
            print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}\n")

        results.append({
            "split": i+1,
            "train_size": len(train_index),
            "test_size": len(test_index),
            "mae": mae,
            "rmse": rmse,
            "y_test": y_test,
            "y_pred": y_pred
        })

    return results

def print_score_array(result, key):
    values = [r[key] for r in results]
    string = "[" + ", ".join(f"{v:.2f}" for v in values) +  "]"
    print(f"{key.upper()} Values: {string} | Mean: {np.mean(values):.2f}")
```

#### Predict Always Mean (MAE: 8.86, RMSE: 13.73)

```python
from sklearn.base import BaseEstimator, RegressorMixin

class MeanBaseline(BaseEstimator, RegressorMixin):
    def fit(self, X, y):
        self.mean_ = y.mean()
        return self

    def predict(self, X):
        return np.full(shape=(len(X),), fill_value=self.mean_)

X = df_trans[['PM10']]
y = df_trans['PM10']

baseline_pipeline = MeanBaseline()
results = evaluate_timeseries_pipeline(X, y, baseline_pipeline, n_splits=5, print_evaluation=False)

print_score_array(results, "mae")
print_score_array(results, "rmse")
```
