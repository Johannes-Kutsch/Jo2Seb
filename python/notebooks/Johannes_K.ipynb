````python

from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor

import pandas as pd
import numpy as np

from ruhken_utils import explore_ts
from Pipeline.DFLoader import DFLoader
from Pipeline.Decomposer.DecomposerFactory import DecomposerFactory
from Pipeline.Decomposer.DecomposerOption import DecomposerOption

from Pipeline.Decomposer.MSTLDecomposer import MSTLDecomposer
from Pipeline.PipelienFactory import PipelineFactory
from Pipeline.Regressor.NaiveForcastRegressor import NaiveForcastRegressor
from Pipeline.Regressor.Utils import create_transformed_target_regressor
from Pipeline.Transformers.ColumnFunctionTransformer import ColumnFunctionTransformer
from Pipeline.ModelWrapper.Recomposer.RecomposerOption import RecomposerOption
from Evaluation.EvaluationMethods import evaluate_timeseries_pipeline, plot_prediction
from Evaluation.Utils import DAILY_EXPERIMENT_NAME

from statsmodels.tsa.seasonal import MSTL
````

````python
df_daily = DFLoader.load_daily_df()
````

# EDA

#### Check if Dates are continuous

````python
df_trans = df_daily.sort_values('date').reset_index(drop=True)
date_diff = df_trans['date'].diff()
print(f"Distance between dates are: {date_diff.unique()[1:]}")
````

#### daily

````python
mstl = MSTL(df_daily['PM10'], periods=[365])
decomp = mstl.fit()
#decomp.resid
explore_ts(df_daily['PM10'], decomp.trend, decomp.seasonal, decomp.resid, lags=100, plot_decomp_separately=False, x_label="Date", y_label="PM10 [μg/m3]")
````

#### log1p PM10 daily

````python
log_PM10 = pd.DataFrame(np.log1p(df_daily[['PM10']]), columns=["PM10"])
mstl = MSTL(log_PM10, periods=[7, 365])
decomp = mstl.fit()

explore_ts(log_PM10, decomp.trend, decomp.seasonal, decomp.resid, lags=100, plot_decomp_separately=False, x_label="Date", y_label="PM10 [μg/m3]")
````

# Models

### Baseline Models

#### Predict Always Mean

````python
mean_pipeline = Pipeline([
    ("predict_mean", DummyRegressor(strategy="mean")),
])

pred_dummy_mean = evaluate_timeseries_pipeline(X=df_daily[['PM10']], y=df_daily['PM10'], pipeline=mean_pipeline,
                                               experiment=DAILY_EXPERIMENT_NAME, model_name="Baseline_Mean", walk_forward=True)
````

````python
plot_prediction(pred_dummy_mean, "Dummy Mean")
````

#### Persistence / Naive Forcast / PM10(t+1) = PM10(t)

````python
naive_forcast_pipeline = Pipeline([
    ("predict_T+1", NaiveForcastRegressor())
])

pred_naive = evaluate_timeseries_pipeline(df_daily[['PM10']], df_daily['PM10'], naive_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_Forcast (T+1)", walk_forward=True)
#pred_hourly = evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, model_name="Baseline_T+1")
````

````python
plot_prediction(pred_naive, "Naive T+1")
````

### Naive MSTL

````python
params = {
    "seasonal_periods": [7, 365],
    "trend_model": LinearRegression(),
    "resid_model": DummyRegressor(strategy="mean"),
}

naive_mstl_forcast_pipeline = Pipeline([
    ("log1p_transform_PM10", ColumnFunctionTransformer(["PM10"])),
    ("Naive_MSTL_Regressor", create_transformed_target_regressor(MSTLDecomposer(seasonal_periods=params["seasonal_periods"], trend_model=params["trend_model"], resid_model= params["resid_model"])))
])

pred_mstl_naive = evaluate_timeseries_pipeline(df_daily[['PM10']], df_daily['PM10'], naive_mstl_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_MSTL_Forcast (T+1)", walk_forward=True, params=params)
````

````python
plot_prediction(pred_mstl_naive, "Naive MSTL")
````

### Prophet + GradientBoosting + Decomposing

````python
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
model_name_prophet_gradiant = "TEST"

pipe = PipelineFactory.create_model_pipeline(model, decomposer_option=DecomposerOption.PROPHET, recomposer_option=RecomposerOption.SEASONAL_TREND)

evaluation_result = evaluate_timeseries_pipeline(df_daily, df_daily['PM10'], pipe, DAILY_EXPERIMENT_NAME, model_name=model_name_prophet_gradiant, walk_forward=False)
````

````python
plot_prediction(evaluation_result, model_name_prophet_gradiant)
````
