## Find & Export Hamburg Port Anchorages

```python
anchorages_df = pandas.read_csv(FilePaths.GFW / "anchorages.csv")

def filter_df_by_coordinates(df, lat_min, lat_max, lon_min, lon_max):
    return df[(df['lat'] >= lat_min) & (df['lat'] <= lat_max) & \
           (df['lon'] >= lon_min) & (df['lon'] <= lon_max)]

def filter_by_label(df, label):
    return df[df['label'].str.strip().str.lower() == label.strip().lower()]
```

## API

```python
url = "https://gateway.api.globalfishingwatch.org/v3/events?limit=1000000&offset=0"

headers = {
    "Authorization": f"Bearer {GFW_TOKEN}",
    "Content-Type": "application/json"
}

body = {
    "datasets": ["public-global-port-visits-events:latest"],
    "geometry": {
        "type": "Polygon",
        "coordinates": [[
            [9.78, 53.45],
            [10.25, 53.45],
            [10.25, 53.74],
            [9.78, 53.74],
            [9.78, 53.45]
        ]]
    },
}

r = requests.post(url, headers=headers, json=body)
data = r.json()

df = pd.DataFrame(data.get("entries", []))
```

```python
df = ru.helper_functions.expand_dic_column(df, 'position')
df = ru.helper_functions.expand_dic_column(df, 'vessel')
df = ru.helper_functions.expand_dic_column(df, 'port_visit')
df = ru.helper_functions.expand_dic_column(df, 'regions')
df = ru.helper_functions.expand_dic_column(df, 'distances')
df = df.drop(['position', 'vessel', 'port_visit', 'regions', 'distances'], axis=1)
df.to_csv(FilePaths.GFW / "low_confidence_port_events.csv")
```

```python
df_port_events = pd.read_csv(FilePaths.GFW / "low_confidence_port_events.csv", index_col=0)
df_port_events["start"] = pd.to_datetime(df_port_events["start"])
df_port_events["end"] = pd.to_datetime(df_port_events["end"])

ru.explore_features(df_port_events, max_print_len=300)
print()

min_date = df_port_events["start"].min()
max_date = df_port_events["start"].max()
delta = max_date - min_date
print(f"Average Events per Day {len(df_port_events) / delta.days}")
```
