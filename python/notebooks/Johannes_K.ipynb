## Find & Export Hamburg Port Anchorages

```python
anchorages_df = pandas.read_csv(FilePaths.GFW / "anchorages.csv")

def filter_df_by_coordinates(df, lat_min, lat_max, lon_min, lon_max):
    return df[(df['lat'] >= lat_min) & (df['lat'] <= lat_max) & \
           (df['lon'] >= lon_min) & (df['lon'] <= lon_max)]

def filter_by_label(df, label):
    return df[df['label'].str.strip().str.lower() == label.strip().lower()]
```

```python
lat_min, lat_max = 53.45, 53.74
lon_min, lon_max = 9.78, 10.25

# Anchorage Feature explanation: https://globalfishingwatch.org/data-download/datasets/public-anchorages%3Av20200316#schema

#anchorages_df_filtered = filter_df_by_coordinates(anchorages_df, lat_min, lat_max, lon_min, lon_max)
anchorages_df_filtered = filter_by_label(anchorages_df, "HAMBURG")

anchorages_df_filtered = anchorages_df_filtered[anchorages_df_filtered["dock"] == True]
anchorages_df_filtered.to_csv(FilePaths.GFW / "anchorages_hamburg.csv")
```

```python
ru.explore_features(anchorages_df_filtered, plot_nan_heatmap=False)
```

```python
ru.eda.Plotting.plot_coordinates(anchorages_df_filtered, color_feature='label_source')
```
