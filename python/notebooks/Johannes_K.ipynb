#### Preprocess and fill missing Values

```python
df = pd.read_csv(FilePaths.UWB_DATA / "Station_857_2018-01-01--2025-12-31.csv")
components_meta = pd.read_csv(FilePaths.UWB_METADATA / "UWB_components_metadata.csv")
```

```python
def get_component_metadata(component_id: int):
    return components_meta.loc[components_meta["component_id"] == component_id].squeeze()

def interpolate_linear(X):
    X_interp = pd.DataFrame(X).interpolate(method='linear', limit_direction='both')
    return X_interp.values

interpolator = FunctionTransformer(interpolate_linear)

component_columns = ["NO2", "SO2", "PM10"]

component_nan_interpolator = ColumnTransformer(
    transformers=[
        ("interpolate_nan", interpolator, component_columns)
    ],
    remainder="drop"
)
```

```python
preprocess_pipe = Pipeline([
    ("rename_features", ru.FeatureRenameTransformer({
        "5": get_component_metadata(5)["component_code"],
        "4": get_component_metadata(4)["component_code"],
        "1": get_component_metadata(1)["component_code"],
        "date_start": "date",
    })),
    ("date_type", ru.ValueTypeCastDatetimeTransformer("date")),
])

df_trans = preprocess_pipe.fit_transform(df)

component_pip = Pipeline([
    ("interpolate_columns", component_nan_interpolator)
])

df_trans[component_columns] = pd.DataFrame(component_pip.fit_transform(df_trans), columns=component_columns)
df_trans_daily = df_trans.set_index("date").resample("D").agg({
    "PM10": "mean"
})
df_trans_daily["date"] = pd.to_datetime(df_trans_daily.index.date)
```

#### Eval Function

```python
def evaluate_timeseries_pipeline(X, y, pipeline, experiment:str, n_splits=5, data_size:float = 1, model_name = "", params = None):
    """
    Evaluates a scikit-learn pipeline on a time series using TimeSeriesSplit.

    Parameters:
    - X: pd.DataFrame, Feature matrix
    - y: pd.Series, Target variable
    - pipeline: sklearn Pipeline or estimator with fit/predict
    - n_splits: int, number of splits for TimeSeriesSplit
    - model_name: str, name of the model

    Returns:
    - results: list of dicts with split info and metrics
    """
    if params is None:
        params = {}

    row_count = int(len(y) * data_size)
    y = y.iloc[:row_count]
    X = X.iloc[:row_count]

    params["rows"] = row_count
    params["features"] = X.columns.tolist()

    mlflow.set_experiment(experiment)

    with mlflow.start_run(run_name=f"{model_name}_cv{n_splits}"):
        time_series_splits = TimeSeriesSplit(n_splits=n_splits)
        results = []

        for i, (train_index, test_index) in enumerate(time_series_splits.split(X)):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            pipeline.fit(X_train, y_train)

            y_pred = pipeline.predict(X_test)
            pred_nan_mask = ~np.isnan(y_pred)

            y_pred = y_pred[pred_nan_mask]
            y_test = y_test.values[pred_nan_mask]

            mae = mean_absolute_error(y_test, y_pred)
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))

            results.append({
                "mae": mae,
                "rmse": rmse,
            })

            mlflow.log_metric(f"mae_cv_{i+1}", mae)
            mlflow.log_metric(f"rmse_cv_{i+1}", rmse)

            print(f"Finished cv_{i+1} | MAE: {mae}, RMSE: {rmse}")

        signature = infer_signature(X_test[pred_nan_mask], y_pred)

        mlflow.log_params(params)

        mlflow.log_metric("avg_mae", get_avg_score(results, "mae"))
        mlflow.log_metric("avg_rmse", get_avg_score(results, "rmse"))
        mlflow.sklearn.log_model(sk_model=pipeline, name=model_name, signature=signature)

def get_avg_score(results, key) -> float:
    return float(np.mean([r[key] for r in results]))
```

### Baseline Models

#### Predict Always Mean

```python
class MeanBaseline(BaseEstimator, RegressorMixin):
    def fit(self, X, y):
        self.mean_ = y.mean()
        return self

    def predict(self, X):
        return np.full(shape=(len(X),), fill_value=self.mean_)

baseline_pipeline = MeanBaseline()
evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], baseline_pipeline, daily_experiment, n_splits=5, model_name="Baseline_Mean")
evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, n_splits=5, model_name="Baseline_Mean")
```

#### Persistence / Naive Forecast / PM10(t+1) = PM10(t)

```python
import numpy as np
from sklearn.base import BaseEstimator, RegressorMixin

class PersistenceBaseline(BaseEstimator, RegressorMixin):
    def fit(self, X, y):
        return self

    def predict(self, X):
        last_values = X['PM10'].shift(1)
        return last_values.values

baseline_pipeline = PersistenceBaseline()
evaluate_timeseries_pipeline(df_trans_daily[['PM10']], df_trans_daily['PM10'], baseline_pipeline, daily_experiment, n_splits=5, model_name="Baseline_T+1")
evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, n_splits=5, model_name="Baseline_T+1")
```

### Temporal Features

#### MSTLARIMA

```python
class MSTLARIMABaseline(BaseEstimator, RegressorMixin):

    def __init__(self, seasonal_periods=(24,168), arima_order=(1,0,1), arima_exog=None):
        self.seasonal_periods = seasonal_periods
        self.arima_order = arima_order
        self.arima_exog = arima_exog

    def fit(self, X, y):

        self.mstl_ = MSTL(y, periods=self.seasonal_periods)
        self.decomp_ = self.mstl_.fit()

        resid = self.decomp_.resid

        self.arima_ = ARIMA(resid, order=self.arima_order, exog=self.arima_exog).fit()

        self.trend_last_ = self.decomp_.trend.iloc[-1]

        seasonal = self.decomp_.seasonal

        if hasattr(seasonal, "columns"):
            self.seasonal_pattern_ = seasonal.sum(axis=1).values
        else:
            self.seasonal_pattern_ = seasonal.values

        return self


    def predict(self, X):

        steps = len(X)

        resid_forecast = self.arima_.forecast(steps)

        trend_forecast = np.repeat(self.trend_last_, steps)

        seasonal_forecast = np.tile(
            self.seasonal_pattern_,
            int(np.ceil(steps / len(self.seasonal_pattern_)))
        )[:steps]

        return trend_forecast + seasonal_forecast + resid_forecast


X = df_trans_daily[['PM10']]
X["is_new_years_eve"] = ((X.index.month == 12) & (X.index.day == 31)).astype(int)
y = df_trans_daily['PM10']

mstl_model = MSTLARIMABaseline(
    seasonal_periods=(7, 365),
    arima_order=(1, 0, 1),
    arima_exog=X[["is_new_years_eve"]]
)

evaluate_timeseries_pipeline(
    X,
    y,
    mstl_model,
    daily_experiment,
    n_splits=5,
    data_size=1,
    model_name="MSTL_ARIMA",
    params={"seasonal_periods": mstl_model.seasonal_periods, "arima_order": mstl_model.arima_order}
)
```
