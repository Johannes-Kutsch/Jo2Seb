# Models

### Baseline Models

#### Predict Always Mean

````python
mean_pipeline = Pipeline([
    ("predict_mean", DummyRegressor(strategy="mean")),
])

pred_dummy_mean = evaluate_timeseries_pipeline(X=df_daily[['PM10']], y=df_daily['PM10'], pipeline=mean_pipeline,
                                               experiment=DAILY_EXPERIMENT_NAME, model_name="Baseline_Mean", walk_forward=True)
````

````python
plot_prediction(pred_dummy_mean, "Dummy Mean")
````

#### Persistence / Naive Forcast / PM10(t+1) = PM10(t)

````python
naive_forcast_pipeline = Pipeline([
    ("predict_T+1", NaiveForcastRegressor())
])

pred_naive = evaluate_timeseries_pipeline(df_daily[['PM10']], df_daily['PM10'], naive_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_Forcast (T+1)", walk_forward=True)
#pred_hourly = evaluate_timeseries_pipeline(df_trans[['PM10']], df_trans['PM10'], baseline_pipeline, hourly_experiment, model_name="Baseline_T+1")
````

````python
plot_prediction(pred_naive, "Naive T+1")
````

### Naive MSTL

````python
params = {
    "seasonal_periods": [7, 365],
    "trend_model": LinearRegression(),
    "resid_model": DummyRegressor(strategy="mean"),
}

naive_mstl_forcast_pipeline = Pipeline([
    ("log1p_transform_PM10", ColumnFunctionTransformer(["PM10"])),
    ("Naive_MSTL_Regressor", create_transformed_target_regressor(MSTLDecomposer(seasonal_periods=params["seasonal_periods"], trend_model=params["trend_model"], resid_model= params["resid_model"])))
])

pred_mstl_naive = evaluate_timeseries_pipeline(df_daily[['PM10']], df_daily['PM10'], naive_mstl_forcast_pipeline, DAILY_EXPERIMENT_NAME, model_name="Naive_MSTL_Forcast (T+1)", walk_forward=True, params=params)
````

````python
plot_prediction(pred_mstl_naive, "Naive MSTL")
````

### Prophet + GradientBoosting + Decomposing

````python
model = GradientBoostingRegressor(n_estimators=100, max_depth=5, learning_rate=0.05, random_state=42)
#model = XGBRegressor(n_estimators=300, max_depth=6, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8,random_state=42, n_jobs=-1)

model_name_prophet_gradiant = "GradientBoost (Prophet, Recompose)"

pipe = PipelineFactory.create_model_pipeline(model, decomposer_option=DecomposerOption.PROPHET, recomposer_option=RecomposerOption.SEASONAL_TREND)

evaluation_result = evaluate_timeseries_pipeline(df_daily, df_daily['PM10'], pipe, DAILY_EXPERIMENT_NAME, model_name=model_name_prophet_gradiant, walk_forward=False)
````

````python
plot_prediction(evaluation_result, model_name_prophet_gradiant)
````
